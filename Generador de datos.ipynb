### Generador de Datos
Primero para generar la tabla ``peliculas``

```python
import pandas as pd
import numpy as np
import random as rd
```

```python
data = pd.read_table("datos/dataMovies.csv", sep=",")
generos = ["Ciencia Ficción", "Horror", "Comedia", "Suspenso", "Animada", "Aventura"]
categorias = ["G","PG","PG13", "R", "NC17"]
notas = [round(2+8*np.random.random(),1) for i in range(5)]
duracion = ["0"+str(1+np.random.randint(3))+":"+str(10+np.random.randint(50)) for i in range(5)]
data_head = data.loc[2:6,]
titulos = list(data_head["Film"])
año = list(data_head["Year"])
data_head
```

```python
to_data_frame = {"titulos": titulos,
                 "puntuacion": notas,
                 "clasificacion": categorias,
                 "duracion": duracion,
                 "año": año}
df = pd.DataFrame(to_data_frame)
# Pasar a dataframe
#df.to_csv('peliculas.csv', index=False)
df
```

Ahora para generar la tabla ``Proveedores``

```python
proveedores = ["Disney+", "Netflix", "Prime Video", "Star+"]
costo = [np.random.randint(low=40, high=70)*100 for i in range(len(proveedores))]

nom_proveedores = {'nombreProveedor': proveedores, 
                  'costo': costo}
df_proveedores = pd.DataFrame(nom_proveedores)
df_proveedores.to_csv('TablaProveedores.csv', index=False)
```

Se genera tabla de ``Link Pelicula Proveedor``

```python
id_proveedor = list(range(1,len(df_proveedores)+1))
id_pelicula = list(range(1,len(df)+1))
def associate_random(id_proveedor, id_pelicula):
    lista_final = list()
    for prov in id_proveedor:
        muestras = rd.randint(1,len(id_pelicula)-1)
        pelis = random.sample(id_pelicula, muestras)
        lista_final += [(id_peli, prov) for id_peli in pelis]
    return lista_final

to_df = associate_random(id_proveedor, id_pelicula)
df_mixpeliprove = pd.DataFrame(to_df, columns=["idProveedor", "idPelicula"])
df_mixpeliprove.to_csv("tablaProveedorPelicula.csv", index=False)
```

Generar la tabla ``Series``

```python
dataSeries = pd.read_table('datos/datosSeries.csv', sep=",")
dataSeries = dataSeries[['Title', 'Rating']].head()
title_serie = list(dataSeries['Title'])
nota_serie = list(dataSeries['Rating'])
año_serie = [2000 + rd.randint(1,20) for i in range(5)]
# categorias -> Sigue igual
temporadas = [rd.randint(1,15) for _ in range(5)]
data_series = pd.DataFrame({'titulo': title_serie, 'Año': año_serie, 'puntuacion': nota_serie, 'clasificacion': categorias, 
                           'temporadas': temporadas})
data_series.to_csv('TablaSeries.csv', index = False)
data_series
```

```python
#rd.choice(['true', 'false'])
def generate_genders(id_peliculas, generos):
    lista_final = list()
    for peli in id_peliculas:
        cantidad_generos = rd.randint(2,3)
        generos_temp = rd.sample(generos, cantidad_generos)
        temporal = [[peli, genero] for genero in generos_temp]
        principal = True
        for temp in temporal:
            if principal:
                temp.append("true")
                principal = False
            else:
                temp.append("false")
            lista_final.append(tuple(temp))
        principal = True
    return lista_final

id_peliculas = list(range(1,6))
df_series = pd.DataFrame(generate_genders(id_peliculas, generos), columns=["idPelicula", "Genero", "EsPrincipal"])
df_series.to_csv("TablaGeneroPelicula.csv", index=False)
```
